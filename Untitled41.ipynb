```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder

# Load your dataset
df = pd.read_csv("/content/kidney_disease.csv")

# List of categorical columns
categorical_cols = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification']

# Apply Label Encoding for categorical columns
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col].astype(str))  # Convert NaN to string before encoding
```

```python
import pandas as pd
import numpy as np



# Replace '?' and tab characters ('\t') with NaN
df.replace(r'^\s*\?$', np.nan, regex=True, inplace=True)

# Trim spaces from column names and values
df.columns = df.columns.str.strip()
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Convert all columns to proper data types
df = df.apply(pd.to_numeric, errors='ignore')

# Check missing values
print(df.isnull().sum())
```

```python


# Apply KNN Imputation
imputer = KNNImputer(n_neighbors=5)  # Using 5 nearest neighbors
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Round categorical columns back to integer values
for col in categorical_cols:
    df_imputed[col] = df_imputed[col].round().astype(int)

# Print to check imputed values
print(df_imputed.head())

# Check if all missing values are filled
print(df_imputed.isnull().sum())  # Should print all zeros
```
